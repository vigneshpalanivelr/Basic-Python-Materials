```python
# Recursive with Memoization
def fibonacci(n):
    if n in fibonacci_chache:
        return fibonacci_chache[n]
    elif n<=1:
        return n
    else:
        fibonacci_chache[n] = fibonacci(n-1) + fibonacci(n-2)
        return(fibonacci_chache[n])

fibonacci_chache = {}

for i in range(0, int(input('Enter a number, N, N>=2 : '))):
    print(i, "\t: ", fibonacci(i))
```

```python
"""
Recursive with Memoization
"""

import functools 

@functools.lru_cache(maxsize = 1000)
def fibonacci(n):    
    if n<=1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))

for i in range(0, int(input('Enter a number, N, N>=2 : '))):
    print(i, "\t: ", fibonacci(i))
```

# Prime Number

```python
"""
Program to find whether a Number is Prime or Not - Using Nested For Loop
"""

Start = int(input("Enter START number : "))
End   = int(input("Enter END   number : "))
Prime = []

for i in range(Start, End+1):
    for j in range(2, i):
        if i%j == 0:
            break
    else:
        Prime.append(i)


print("The count of PN is :", len(Prime))
print("The Prime Numbers  :" , Prime)
```

```python
"""
Program to find whether a Number is Prime or Not - Using Nested For Loop
"""

Start = int(input("Enter START number : "))
End   = int(input("Enter END   number : "))
Prime = []

for i in range(Start, End+1):
    for j in range(2, i):
        if i%j == 0:
            break
        if j*j > i:
            continue
    else:
        Prime.append(i)

print("The count of PN is :", len(Prime))
print("The Prime Numbers  :" , Prime)
```

```python
"""
Program to find whether a Number is Prime or Not - Using While Loop 
"""

Start = int(input("Enter START number : "))
End   = int(input("Enter END   number : "))
Prime = []

def prime(i, End):
    while True:
        if  i > End:
            break
        for j in range(2, i):
            if i%j == 0:
                break
            if j*j > i:
                continue
        else:
            Prime.append(i)
        i += 1

prime(Start, End)
print(Prime)
```

```python
"""
Program to find whether a Number is Prime or Not - Using Recursion 
"""

Start = int(input("Enter START number : "))
End   = int(input("Enter END   number : "))
Prime = []

def isPrime(i, n = 2, End=None): 
	for n in range(2, i):
		if (i <= 2):
			Prime.append(i) if(i == 2) else False
		elif (i % n == 0): 
			break
		if (n * n > i):
			continue
	else:
		Prime.append(i)

	# Check for next divisor
	if i < End:
		isPrime(i + 1, End = End)
	else:
		print(Prime)
isPrime(Start, End=End)
```

# Palindrome

```python
"""
Program to find whether a String is Palindrome or Not - Using list Reverse
"""

string = input(("Enter a string: "))

if(string==string[::-1]):
      print("The string is a palindrome")
else:
      print("Not a palindrome")
```

```python
"""
Program to find whether a String is Palindrome or Not - Using While loop
"""

string = input(("Enter a string: "))
start  = 0
end    = len(string) - 1

while True:
    if start == end :
        print("The string is a palindrome")
        break
    if string[start] == string[end]:
        start += 1
        end   -= 1
    else:
        print("The string is NOT a palindrome")
        break
```

#  Power of a number - Recursion

```python
def recur(b, p):
    return b * recur(b, p-1) if p !=0 else 1

print(recur(int(input("Enter the Base: ")), int(input("Enter the Power: "))))
```

# Digits in Reverse Order 

```python
num = int(input("Enter a number with multiple digit: "))

while num>0:
    x   = num % 10
    num = (num - x)//10
    print(x, end='')
```

```python
num = int(input("Enter a number with multiple digit: "))
rev_num = 0

while(num) :
    rem     = num % 10
    rev_num = rev_num * 10 + rem
    num //= 10
    
print('Reverse number is: ', rev_num)
```

# BMI Calculator

```python
height = float(input("Enter height in CMs: "))
weight = float(input("Enter weight in kg: "))

BMI = ( weight * 100 * 100 ) / (height ** 2)

print("Your BMI is: {0} and you are: ".format(BMI), end='')

#conditions
if ( BMI < 16):
    print("severely underweight")

elif ( BMI >= 16 and BMI < 18.5):
    print("underweight")

elif ( BMI >= 18.5 and BMI < 25):
    print("Healthy")

elif ( BMI >= 25 and BMI < 30):
    print("overweight")

elif ( BMI >=30):
    print("severely overweight")
```
